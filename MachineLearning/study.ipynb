```python
from sklearn import linear_model
import matplotlib.pyplot as plt
reg = linear_model.LinearRegression()
reg.fit([[0, 0], [1, 1], [2, 2]], [0, 1, 2])
reg.coef_
plt.plot()
plt.show()
```
