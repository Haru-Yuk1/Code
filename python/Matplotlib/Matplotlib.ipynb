## Matplotlib Pyplot





```python
import matplotlib.pyplot as plt
import numpy as np
```

以下是一些常用的 pyplot 函数：\
plot()：用于绘制线图和散点图\
scatter()：用于绘制散点图\
bar()：用于绘制垂直条形图和水平条形图\
hist()：用于绘制直方图\
pie()：用于绘制饼图\
imshow()：用于绘制图像\
subplots()：用于创建子图

```python

plt.plot([1, 2, 3, 4])
plt.ylabel('some numbers')
plt.show()
```

```python

x = np.arange(0,4*np.pi,0.1) # start,stop,step
y = np.sin(x)
z = np.cos(x)
plt.plot(x,y,x,z)
plt.show()
```

```python

x = np.array([1, 2, 3, 4, 5, 6, 7, 8])
y = np.array([1, 4, 9, 16, 7, 11, 23, 18])

plt.scatter(x, y)
plt.title('scatter')
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

#plot 1:
xpoints = np.array([0, 6])
ypoints = np.array([0, 100])

plt.subplot(1, 2, 1)
plt.plot(xpoints,ypoints)
plt.title("plot 1")

#plot 2:
x = np.array([1, 2, 3, 4])
y = np.array([1, 4, 9, 16])

plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.title("plot 2")

plt.suptitle("RUNOOB subplot Test")
plt.show()
```
